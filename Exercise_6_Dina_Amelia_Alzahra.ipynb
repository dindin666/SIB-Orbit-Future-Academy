2. Buat sebuah program OOP dengan ketentuan:

  - berbeda dari contoh program yang ada di materi (ppt dan colab)
  - memiliki 1 parent class
  - memiliki setidaknya 2 child class
  - menerapkan konsep dasar OOP (kecuali abstraction)
  - dilengkapi dengan docstring, contoh docstring bisa dilihat dalam program hero.py

  Setelah program selesai, buatlah:
  - setidaknya 2 instance untuk setiap child class
  - Class Diagram-nya

```python

class furnitur:
  """Bantuan Library Furnitur:
  NAME
    furnitur

  CLASSES
    kursi
    lemari
  """
  nama = ''
  warna = ''
  berat = ''

  def __init__(self, nama, warna, berat):
    self.nama = nama
    self.warna = warna
    self.berat = berat 

  def dibuat(self):
    print('Dibuat Oleh Tukang')

  def dijual(self):
    print('Dijual di Toko Meubel')

  def get_info(self):
    return self.nama, self.warna, self.berat

class kursi(furnitur): 
  nama = ''
  material = ''

  def __init__(self, nama, material):
    self.nama = nama
    self.material = material

  def dibuat(self):
    print(f'{self.nama} dibuat dengan material{self.material}')

  def dijual(self):
    print('Dijual di Toko Meubel')

  def dibeli(self):
    print('Dibeli oleh Konsumen')
  
  def get_info1(self):
    print('Nama \t:', self.nama)
    print('Warna \t:', self.warna)
    print('Berat \t:', self.berat)
    print('Material:', self.material)
    
class lemari(furnitur): 
  nama = ''
  jumlah_pintu = ''

  def __init__(self, nama, jumlah_pintu):
    self.nama = nama
    self.jumlah_pintu = jumlah_pintu

  def dibuat(self):
    print(f'{self.nama} dibuat dengan total pintu{self.jumlah_pintu}')

  def dijual(self):
    print('Dijual di Toko Meubel')

  def dibeli(self):
    print('Dibeli oleh Konsumen')
  
  def get_info2(self):
    print('Nama\t :', self.nama)
    print('Warna\t :', self.warna)
    print('Berat \t :', self.berat)
    print('Jumlah Pintu :', self.jumlah_pintu)
```

```python
sofa = kursi('nama', 'material')
sofa.nama = 'Sofa'
sofa.warna = 'Abu'
sofa.berat = 9
sofa.material = 'katun'
sofa.get_info1()
```

```python
wing_chair = kursi('nama', 'material')
sofa.nama = 'Wing Chair'
sofa.warna = 'Gold'
sofa.berat = 7
sofa.material = 'Linen'
sofa.get_info1()
```

```python
drawer = lemari('nama', 'jumlah_pintu')
lemari.nama = 'Lemari Drawer'
lemari.warna = 'Putih'
lemari.berat = 20
lemari.jumlah_pintu = 3
drawer.get_info2()
```
